## popularity based Recommender system

- display the top 50 books with highest average rating but consider only those books which has min 250 ratings.

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
# counting the number of ratings for each book
# and renaming the column to 'num_ratings'
num_ratings_df = ratings_with_name.groupby(
    'Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
```

```python
num_ratings_df.head()
```

```python
# average rating of each book
avg_rating_df = ratings_with_name.groupby(
    'Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
```

```python
avg_rating_df.head()
```

```python
# Merging the num_ratings_df with ratings_with_name to get the number of ratings for each book
popular_df = num_ratings_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
# keeping only those books which have more than 250 ratings
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values(
    'num_ratings', ascending=False).reset_index(drop=True)
```

```python
# keeping only the top 100 popular books
popular_df = popular_df.head(100)
```

```python
popular_df
```

```python
# removing duplicates based on 'Book-Title' and
# keeping only the relevant columns for the final output
# merging with books dataframe to get book details

popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates("Book-Title")[
    ["Book-Title", "Book-Author", "Year-Of-Publication", "num_ratings", "avg_rating", "Image-URL-M"]]
```

## Collaborative Filtering Based Recommender System

- considering only those users (knowledgable users) who have rated more than 200 books.
- considering only those books (famous books) who has rated atleast 50 times by knowledgable users.


```python
ratings_with_name.head()
```

```python
# considering users who have rated more than 200 books

x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200

# boolean indexing to get the User-IDs of those users
knowledgable_users = x[x].index
knowledgable_users
```

```python
# filtering ratings to include only those from knowledgable users

filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(
    knowledgable_users)]
# we have done filtering based on knowledgable users
```

```python
# we have done filtering based on books who have been rated at least 50 times

y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50

famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(
    famous_books)]
# we have done filtering based on knowledgable users
```

```python
# converting "Book-Title", "Book-Author" columns to lowercase
final_ratings["Book-Title"] = final_ratings["Book-Title"].str.lower()
final_ratings["Book-Author"] = final_ratings["Book-Author"].str.lower()
```

```python
final_ratings.drop_duplicates()
```

```python
# applying pivot table to get the final ratings matrix
final_ratings_table = final_ratings.pivot_table(
    index="Book-Title", columns='User-ID', values='Book-Rating')
```

```python
final_ratings_table.fillna(0, inplace=True)
```

```python
# converting "Book-Title" column to lowercase
final_ratings_table.index = final_ratings_table.index.str.lower()
```

```python
final_ratings_table
```

```python
final_ratings_table.to_csv('final_ratings_table.csv')
```

## model training


```python
from sklearn.metrics.pairwise import cosine_similarity
# model training
similarity_score = cosine_similarity(final_ratings_table)
```

```python
similarity_score.shape
```

```python
# This will return the index of the book "1984"
np.where(final_ratings_table.index == "1984")[0][0]
```

```python
def recommend(book_name):
    # fetch the index of the book
    book_index = np.where(final_ratings_table.index == book_name)[0][0]

    # find the similarity score of the book with all other books
    # and display the top 7 most similar books
    similar_books = sorted(list(enumerate(
        similarity_score[book_index])), key=lambda x: x[1], reverse=True)[1:8]

    data = []
    # fetching the details of the similar books
    for i in similar_books:
        item = []
        # Use case-insensitive matching to find the book
        temp_df = books[books['Book-Title'].str.lower() == final_ratings_table.index[i[0]]]

        if not temp_df.empty:  # Only add if we found a match
            item.extend(list(temp_df.drop_duplicates(
                "Book-Title")["Book-Title"].values))
            
            item.extend(list(temp_df.drop_duplicates(
                "Book-Title")["Book-Author"].values))
            
            item.extend(list(temp_df.drop_duplicates(
                "Book-Title")["Year-Of-Publication"].values))
            
            item.extend(list(temp_df.drop_duplicates(
                "Book-Title")["Image-URL-M"].values))
            
            data.append(item)

    return data
```

```python
book_name = "zoya"
book_name = book_name.lower()
print(book_name)
recommend(book_name)
```
